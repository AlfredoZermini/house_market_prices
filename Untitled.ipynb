```python
import pandas as pd
from scipy import stats
import numpy as np
import sys
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
np.set_printoptions(threshold=sys.maxsize)
```

```python
# load data
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
# start preparing the training data
# check which columns contain nans
df_train.columns[df_train.isna().any()].tolist()
```

```python
# check number of nans per column
nan_sum = df_train.isna().sum()
nan_sum[0:41]
```

```python
nan_sum[40:]
```

```python
# drop columns containing a large number of nans
df_train_dropcol = df_train.drop(['Alley', 'LotFrontage', 'FireplaceQu','PoolQC','Fence','MiscFeature'], axis = 1)
```

```python
df_train_dropcol
```

```python
nan_sum = df_train_dropcol.isna().sum()
nan_sum[0:41]
```

```python
# drop rows with >= 2 nans
df_train_droprows = df_train_dropcol.dropna(thresh=df_train_dropcol.shape[1]-2)
```

```python
df_train_droprows
```

```python
# fill the few remaining nans with the value on the next row
df_train_nonans = df_train_droprows.ffill(axis=0)
```

```python
# check if any nan is left
df_train_nonans.isnull().values.any()
```

```python
# convert into 1-hot encoding
train_features = pd.get_dummies(df_train_nonans)
```

```python
train_features
```

```python
# delete very extreme outliers by keeping only rows where all feature values are below 10 standard deviations
train_features = train_features[(np.abs(stats.zscore(train_features)) < 10).all(axis=1)]
```

```python
# prepare testing data in a similar way
df_test_dropcol = df_test.drop(['Alley', 'LotFrontage', 'FireplaceQu','PoolQC','Fence','MiscFeature'], axis = 1)
df_test_droprows = df_test_dropcol.dropna(thresh=df_test_dropcol.shape[1]-2)
df_test_nonans = df_test_droprows.ffill(axis=0)
test_features = pd.get_dummies(df_test_nonans)
```

```python
df_test_nonans.isnull().values.any()
```

```python
test_features
```

```python
# train and test features size do not match: this is a problem
a = list(test_features.columns.values)
b = list(train_features.columns.values)
list(set(a)-set(a).intersection(b))
```

```python
# create labels
y_train = np.array(train_features['SalePrice'])
y_train.shape
```

```python
# create input features
train_features_noprice = train_features.drop('SalePrice', axis = 1)
x_train = np.array(train_features_noprice)
x_test = np.array(test_features)
print(x_train.shape, y_train.shape, x_test.shape)
```

```python
# choose rf with 1000 trees (large enough, good training time)
rfr = RandomForestRegressor(n_estimators = 1000)
```

```python
# train
rfr.fit(x_train, y_train)
```

```python
# get predictions on training set
pred = rfr.predict(x_train)
```

```python
# get RMSE
rms = sqrt(mean_squared_error(np.log(y_train), np.log(pred)))
print('RMSE =', rms)
```

```python
# get actual predition on testing set
pred_test = rfr.predict(x_test)
# SIZE DO NOT MATCH: SOME KIND OF PADDING WAS REQUIRED
```
